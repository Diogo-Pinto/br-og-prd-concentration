Análise de Concentração da Produção de Petróleo no Brasil (2000-2025)

1. **Descrição do Projeto**
- Este projeto realiza uma análise da concentração da produção de petróleo no Brasil, com foco na distribuição do volume produzido  entre os diferentes campos produtores ao longo do tempo.
- O objetivo é entender se a produção nacional está concentrada em poucos ativos e como essa concentração evoluiu entre 2000 e 2025.
- Trata-se de um problema de análise exploratória de dados (não supervisionado), cujo foco é identificar padrões e estruturas inerentes aos dados históricos de produção


2. **Objetivos da Análise**
- Identificar Ativos Críticos: Aplicar a Curva de Pareto para identificar quais campos são responsáveis pela maior parcela da produção nacional
- Analisar Participação Relativa: Calcular a participação percentual de cada campo na produção mensal, com foco no domínio dos principais produtores (Top 5)
- Avaliar a Dinâmica Temporal: Observar a evolução da produção e da concentração ao longo do período analisado
- Mensurar a Concentração: Utilizar o índice HHI para quantificar o nível de concentração da produção entre os campos


3. **Hipótese Central**
- A entrada em operação dos campos da camada pré-sal, a partir de 2010, aumentou o nível de concentração da produção de petróleo no Brasil, elevando o índice HHI e ampliando a participação relativa dos principais campos produtores.
- Para fins de interpretação, serão utilizados os seguintes critérios para o HHI:
    - HHI < 0.15 → baixa concentração /d
    - 0.15 ≤ HHI < 0.25 → concentração moderada
    - HHI ≥ 0.25 → alta concentração


4. **Fonte e Escopo dos Dados**
- Conteúdo: Produção mensal de petróleo por campo
- Variável principal: Volume de produção de petróleo (m³)
- Abrangência: Campos onshore e offshore
- Período: Janeiro de 2000 a Dezembro de 2025


```python
# Importação de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações de visualização para o Jupyter/Colab
%matplotlib inline
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 6] # Define um tamanho padrão melhor para os gráficos

# Configura o pandas para mostrar números flutuantes com 2 casas decimais e separador de milhar
pd.options.display.float_format = '{:,.2f}'.format

# Caminho dos dados no GitHub (Versão Raw)
DATA_URL = "https://raw.githubusercontent.com/Diogo-Pinto/br-og-prd-concentration/main/data/raw/prod_2000_2025.csv"

# Carregamento do DataFrame com tratamento de erro básico
try:
    df = pd.read_csv(DATA_URL, sep=',', encoding='utf-8-sig')
    print(f"Sucesso! O dataset possui {df.shape[0]} linhas e {df.shape[1]} colunas.")
except Exception as e:
    print(f"Erro ao carregar os dados: {e}")

# Visualização das primeiras instâncias
df.head()
```

```python
# --- Análise Estatística Descritiva ---
# O objetivo aqui é entender a dispersão dos dados de produção antes de calcular o HHI.

# Gerando o resumo estatístico para as variáveis numéricas
estatisticas = df.describe()

# Adicionando a contagem de valores ausentes (requisito do checklist)
estatisticas.loc['missing_values'] = df.isnull().sum()

display(estatisticas)
```

Insights:

Ao observar o resumo estatístico, identifiquei pontos que exigirão atenção no pré-processamento:

1. **Inconsistências Físicas:** Colunas como `Vol. Gás Disponível` e `Vol. Água Descartada` apresentam valores negativos (ex: -1.625.204), o que é fisicamente impossível para volume produzido. 
2. **Campos Inativos:** O valor mínimo de 0 em `Produção de Petróleo` indica meses sem extração para certos campos, o que possívelmente se deve a períodos de parada programada ou eventos de shutdown. 
3. **Dados Faltantes:** A variação no `count` (ex: Lâmina d'água vs. Gás Injetado) confirma que nem todos os dados estão disponíveis para todos os campos em todos os meses.

Especificamente sobre a variável alvo: Produção de Petróleo (m³)
1. A média (32.924 m3) é fortemente enviesada pelo valor máximo (5.292.606 m3).
2. O desvio padrão (201.161 m3) é quase 6 vezes maior que a média, indicando uma distribuição de cauda longa.
3. A mediana (383 m3) revela que o 'campo comum' no Brasil produz muito pouco perante os gigantes, o que antecipa um HHI de alta concentração.
4. 75% dos registros de produção no Brasil (ao longo dos 25 anos) produzem menos de 4.300 m³ por mês. No entanto, o líder produz 1200 vezes mais que o limite do terceiro quartil.
