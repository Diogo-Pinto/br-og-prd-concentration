## Definição do Problema

A produção de petróleo no Brasil passou por mudanças estruturais relevantes nas últimas décadas, especialmente após a entrada em operação dos campos da camada pré-sal a partir de 2010. Esse movimento levanta a necessidade de compreender como a produção está distribuída entre os diferentes campos produtores e se houve aumento na dependência de um número reduzido de ativos de alta relevância.

Diante desse contexto, o problema central deste estudo consiste em avaliar o grau de concentração da produção de petróleo no Brasil ao longo do tempo, buscando identificar se essa produção está distribuída de forma equilibrada entre diversos campos ou concentrada em poucos ativos dominantes.

Além da mensuração direta da concentração por meio do índice de Herfindahl-Hirschman (HHI) ao longo do tempo, o problema também é abordado sob diferentes perspectivas, considerando características dos campos como profundidade, qualidade do petróleo (grau API) e estágio do ciclo de vida produtivo.

Trata-se de um problema de análise exploratória de dados, cujo objetivo é identificar padrões estruturais na distribuição da produção e compreender sua evolução ao longo do período analisado (2000–2025).

## Hipóteses do Problema
A entrada em operação dos campos da camada pré-sal, a partir de 2010, aumentou o nível de concentração da produção de petróleo no Brasil, elevando o índice HHI e ampliando a participação relativa dos principais campos produtores.

As hipóteses traçadas são as seguintes:
1. Ao longo do tempo, especialmente após a entrada em operação dos campos do pré-sal, houve um aumento no nível de concentração da produção de petróleo no Brasil, refletido por elevação do índice HHI.

2. A produção nacional no período mais recente é predominantemente proveniente de campos em águas ultraprofundas.

3. A produção de petróleo no Brasil no período mais recente está majoritariamente associada a campos que produzem óleo de grau API médio.

4. No período mais recente, a produção de petróleo no Brasil está concentrada em campos classificados como maduros, em termos de ciclo de vida.

Para fins de interpretação, serão utilizados os seguintes critérios:

- **Herfindahl-Hirschman Index (HHI)**  
O Índice de Herfindahl-Hirschman (HHI) é uma métrica amplamente utilizada para medir o nível de concentração em um mercado ou, neste caso, na produção de petróleo entre diferentes campos. O HHI é calculado como a soma dos quadrados das participações relativas de cada campo na produção total em um determinado período. Essa abordagem atribui maior peso aos campos com maior participação, tornando a métrica sensível à dominância de poucos ativos. No contexto deste trabalho, o HHI será utilizado para avaliar o grau de concentração da produção ao longo do tempo, permitindo identificar mudanças estruturais na distribuição da produção entre os campos. 
    - HHI < 0.15 → baixa concentração /d
    - 0.15 ≤ HHI < 0.25 → concentração moderada
    - HHI ≥ 0.25 → alta concentração


- **Qualidade do Petróleo (Grau API)**
O grau API foi discretizado para classificar o tipo de petróleo produzido, com base em sua densidade e valor econômico.Essa segmentação possibilita analisar se a concentração da produção está associada a determinados tipos de petróleo.
    - API < 22 → **Pesado**
    - 22 ≤ API < 31 → **Médio**
    - 31 ≤ API < 45 → **Leve**
    - API ≥ 45 → **Condensado de Gás**


- **Idade do Campo**
A idade do campo foi calculada a partir da diferença entre o ano de produção e o ano de início da produção, sendo posteriormente categorizada em estágios do ciclo de vida produtivo. Essa classificação permite avaliar se a concentração da produção está associada a campos mais recentes (como os do pré-sal) ou a ativos mais antigos.
    - Idade < 5 anos → **Novo**
    - 5 ≤ Idade < 15 anos → **Maduro**
    - Idade ≥ 15 anos → **Antigo**

- **Profundidade (Lâmina d’Água)**
A lâmina d’água foi discretizada para classificar os campos offshore de acordo com sua profundidade operacional. Essa categorização permite analisar a relação entre profundidade e concentração da produção, especialmente considerando o papel dos campos em águas profundas e ultraprofundas no contexto do pré-sal.
    - Lâmina d’água < 200 m → **Raso**
    - 200 ≤ Lâmina d’água < 1000 m → **Profundo**
    - Lâmina d’água ≥ 1000 m → **Ultraprofundo**

### Identificação de Fragmentações na Nomenclatura dos Campos

Durante a análise dos nomes dos campos, foram identificados padrões que indicam a existência de múltiplas denominações para um mesmo ativo produtivo. Essas variações incluem:

- Sufixos associados a regimes contratuais distintos (ex: "_ECO");
- Prefixos relacionados a antecipação de produção (ex: "AnC_");
- Outras variações operacionais ou administrativas.

Essas diferenças não necessariamente representam campos distintos do ponto de vista geológico ou produtivo, podendo refletir aspectos regulatórios, contratuais ou fases de desenvolvimento de um mesmo ativo.

A presença dessas fragmentações pode levar à divisão artificial da produção entre múltiplos registros, distorcendo a análise de concentração ao reduzir indevidamente a participação relativa dos principais campos.

Com base nessa constatação, e após validação com especialista da área, optou-se por **unificar as denominações dos campos**, tratando-os como unidades produtivas únicas. Essa decisão visa garantir que a análise represente de forma mais fiel a concentração da produção entre os ativos, alinhando a modelagem dos dados ao objetivo do estudo.

Ressalta-se que essa abordagem implica uma simplificação da estrutura original dos dados, ao desconsiderar diferenças contratuais e operacionais entre registros associados ao mesmo campo. No entanto, tal simplificação é considerada adequada para os fins desta análise.

Com o objetivo de aprofundar a análise da distribuição da produção ao longo do tempo e complementar as visualizações gráficas, foi construído um DataFrame resumo contendo, para cada ano, estatísticas descritivas baseadas em quantis (p10, p50, p90 e p99), além da quantidade de campos em operação. Esse comportamento é consistente com um possível processo de polarização, no qual a produção tende a se concentrar em um número reduzido de campos altamente produtivos, enquanto uma parcela crescente dos campos apresenta baixa relevância produtiva.

Adicionalmente, a evolução da quantidade de campos ao longo do tempo não parece ser suficiente, isoladamente, para explicar o aumento da dispersão observado. Embora haja crescimento no número de ativos, a queda acentuada nos quantis centrais e inferiores sugere que a dinâmica observada pode estar associada ao envelhecimento dos campos e à coexistência de ativos em diferentes estágios do ciclo de vida, e não apenas a um efeito de expansão do sistema.

Dessa forma, a análise conjunta dos quantis e da quantidade de campos reforça a hipótese de que o setor pode estar passando por um processo de concentração produtiva, no qual poucos campos tendem a responder por uma parcela crescente da produção total. Esses indícios serão explorados de forma mais aprofundada nas etapas seguintes do trabalho.

```python
df_quantis = (
    df_oil_prod
    .groupby('Ano')['Produção de Petróleo (m³)']
    .quantile([0.1, 0.5, 0.9, 0.99])
    .unstack()
    .rename(columns={
        0.1: 'p10',
        0.5: 'p50',
        0.9: 'p90',
        0.99: 'p99'
    })
)

df_campos = (
    df_oil_prod
    .groupby('Ano')['Campo']
    .nunique()
    .rename('qtd_campos')
)

df_resumo = df_quantis.join(df_campos)

df_resumo
```

Um aspecto relevante observado é a ausência de registros no intervalo entre 0 e aproximadamente 4 na escala logarítmica, indicando uma descontinuidade entre campos sem produção e aqueles com produção efetiva. Esse padrão sugere que, uma vez ativos, os campos tendem a operar em níveis de produção significativamente superiores a zero, não havendo concentração relevante em níveis muito baixos de produção.

Adicionalmente, observa-se que a maior densidade de registros se concentra em torno do valor 10 na escala logarítmica, indicando que a maior parte dos campos ativos opera em uma faixa relativamente homogênea de produção intermediária.

Ao mesmo tempo, a presença de uma cauda longa à direita evidencia a existência de valores elevados de produção associados a um número reduzido de campos. ### Padronização dos Campos e Agregação dos Dados

Durante a análise exploratória, foram identificadas inconsistências na representação dos campos produtores, com múltiplas denominações associadas a um mesmo ativo produtivo. Essas variações incluem diferenças contratuais (ex: sufixo "_ECO"), fases de produção (ex: prefixo "AnC_") e distinções geográficas (ex: "NORTE DE", "SUL DE", "SUDOESTE").

Considerando que o objetivo deste estudo é avaliar a concentração da produção sob a perspectiva dos campos como unidades produtivas geográficas, optou-se por padronizar os nomes dos campos, unificando essas variações em uma única identificação, denominada `campo_base`. Essa transformação busca evitar a fragmentação artificial da produção entre múltiplos registros associados ao mesmo ativo.

Após a padronização, foi realizada a agregação dos dados, de modo que cada observação represente a produção de um campo (`campo_base`) em um determinado período (`data`).

**Estratégia de agregação**

A agregação foi realizada utilizando diferentes critérios, de acordo com a natureza de cada variável:

- **Produção de Petróleo (m³)**: foi utilizada a soma dos valores, representando o volume total produzido pelo campo no período.

- **Grau API (ºAPI)**: foi utilizada a mediana, por se tratar de uma propriedade físico-química do petróleo que tende a ser relativamente estável dentro de um campo. A mediana foi escolhida por ser menos sensível a valores extremos ou inconsistentes.

- **Lâmina d’Água**: foi utilizada a mediana, como forma de representar a profundidade típica do campo, evitando distorções causadas por variações pontuais ou registros atípicos.

- **Idade do Campo**: foi utilizada a mediana, garantindo uma medida representativa do estágio de maturidade do campo, especialmente em casos onde múltiplos registros estão associados a diferentes momentos ou condições operacionais.

Essa abordagem permite consolidar os dados em uma estrutura consistente, adequada para o cálculo de métricas de concentração, como a participação relativa e o índice HHI.

**Considerações**

A padronização e agregação dos dados implicam uma simplificação da estrutura original do dataset, ao desconsiderar diferenças contratuais, operacionais e, em alguns casos, espaciais entre registros. No entanto, essa simplificação foi considerada necessária para garantir uma representação mais fiel da concentração da produção entre os principais ativos produtores.

```python
import re
import unicodedata

def normaliza_texto(texto):
    # remove acento
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    return texto

def limpa_nome(campo):
    original = campo
    
    # normaliza acento e caixa
    campo = normaliza_texto(campo.upper())
    
    # remove AnC e ECO
    campo = re.sub(r'^ANC_', '', campo)
    campo = re.sub(r'_ECO$', '', campo)
    
    # remove prefixos
    campo = re.sub(
        r'^(NORTE DE |SUL DE |LESTE DE |OESTE DE |NORDESTE DE |NOROESTE DE |SUDESTE DE |SUDOESTE DE )',
        '',
        campo
    )
    
    # remove sufixos
    campo = re.sub(
        r'( NORTE| SUL| LESTE| OESTE| NORDESTE| NOROESTE| SUDESTE| SUDOESTE)$',
        '',
        campo
    )
    
    return campo.strip()

# aplicar
df_oil_prod_tratado['campo_base'] = df_oil_prod_tratado['Campo'].apply(limpa_nome)
```

**Consolidação Base Anual**

Com o objetivo de tornar a análise mais robusta e reduzir a influência de variações de curto prazo, os dados originalmente mensais foram agregados para uma base anual.

A granularidade mensal, embora mais detalhada, pode refletir flutuações operacionais pontuais, como paradas para manutenção, variações sazonais ou oscilações temporárias na produção, que não necessariamente representam mudanças estruturais na distribuição da produção entre os campos.

Dessa forma, optou-se por consolidar os dados por ano, permitindo uma visão mais estável e representativa da produção de cada campo ao longo do tempo. Essa abordagem é especialmente adequada para o cálculo de métricas de concentração, como o índice HHI, e para a análise comparativa entre diferentes períodos.

Na agregação anual, a produção de petróleo foi somada, representando o volume total produzido por cada campo no ano. Para as demais variáveis (grau API, lâmina d’água e idade), foi utilizada a mediana, por se tratar de uma medida robusta e adequada para representar as características típicas de cada campo ao longo do período.

```python
def anualiza_producao(df):
    df_anual = (
        df
        .assign(Ano=df['data'].dt.year)
        .groupby(['Ano', 'Campo'])
        .agg({
            'Produção de Petróleo (m³)': 'sum',
            'ºAPI': 'median',
            'Lâmina D´Água': 'median',
            'idade': 'median'
        })
        .reset_index()
    )
    
    return df_anual

df_oil_prod_anual = anualiza_producao(df_oil_prod_agg)
```

### Participação Relativa dos Campos

Para avaliar o nível de concentração da produção de petróleo entre os campos, foi calculada a **participação relativa de cada campo na produção total mensal**.

Essa métrica representa a fração da produção total de petróleo, em um determinado mês, atribuída a cada campo produtor. Dessa forma, permite comparar a relevância de diferentes campos independentemente da variação do volume total produzido ao longo do tempo.

A participação foi calculada da seguinte forma:

\[
Participação_{i,t} = \frac{Produção_{i,t}}{\sum_{j} Produção_{j,t}}
\]

Onde:

- \( Produção_{i,t} \) corresponde ao volume produzido pelo campo *i* no período *t*;
- \( \sum_{j} Produção_{j,t} \) representa a produção total de petróleo no período *t*, considerando todos os campos.

Essa abordagem garante que, para cada período, a soma das participações de todos os campos seja igual a 1 (ou 100%), permitindo a aplicação de métricas de concentração, como o índice de Herfindahl-Hirschman (HHI).

Além disso, a utilização de participação relativa permite eliminar o efeito de crescimento absoluto da produção ao longo do tempo, focando exclusivamente na distribuição da produção entre os campos.

```python
# Total por ano
df_oil_prod_anual['total_ano'] = (
    df_oil_prod_anual
    .groupby('Ano')['Produção de Petróleo (m³)']
    .transform('sum')
)

# Participação do campo no ano
df_oil_prod_anual['participacao'] = (
    df_oil_prod_anual['Produção de Petróleo (m³)'] /
    df_oil_prod_anual['total_ano']
)
```

Verifica-se que a média das participações agrupadas por ano é igual a um, validando o cálculo

```python
validacao = df_oil_prod_anual.groupby('Ano')['participacao'].sum()
validacao.mean()
```

### Índice de Herfindahl-Hirschman (HHI)

O Índice de Herfindahl-Hirschman (HHI) é uma métrica amplamente utilizada para medir o nível de concentração em um mercado. Neste estudo, o HHI foi aplicado para avaliar a concentração da produção de petróleo entre os campos produtores ao longo do tempo.

O índice é calculado como a soma dos quadrados das participações relativas de cada campo em um determinado período:

\[
HHI_t = \sum_{i} (Participação_{i,t})^2
\]

Onde:

- \( Participação_{i,t} \) representa a participação do campo *i* na produção total no período *t*.

O uso do quadrado das participações atribui maior peso aos campos com maior relevância produtiva, tornando o índice sensível à dominância de poucos ativos.

Para interpretação dos resultados, foram adotados os seguintes critérios:

- HHI < 0.15 → baixa concentração  
- 0.15 ≤ HHI < 0.25 → concentração moderada  
- HHI ≥ 0.25 → alta concentração  

A análise da evolução temporal do HHI permite identificar mudanças estruturais na distribuição da produção, indicando se o setor se tornou mais ou menos concentrado ao longo do tempo.

```python
# Cálculo do HHI por mês
hhi = (
    df_oil_prod_anual
    .groupby('Ano')['participacao']
    .apply(lambda x: (x**2).sum())
    .reset_index(name='HHI')
)
```

### Feature Engineering

**Classificação do Grau API**

O grau API é uma medida da densidade do petróleo e está diretamente relacionado ao seu valor econômico. Para fins analíticos, essa variável foi discretizada em três categorias:

- API < 22 → **Pesado**
- 22 ≤ API < 31 → **Médio**
- API ≥ 31 → **Leve**

Essa segmentação permite avaliar se a concentração da produção está associada a determinados tipos de petróleo.

```python
def classifica_api(api):
    if api < 22:
        return 'Pesado'
    elif api < 31:
        return 'Médio'
    else:
        return 'Leve'

df_2025['api_cat'] = df_2025['ºAPI'].apply(classifica_api)
```

**Classificação da Idade do Campo**

A idade do campo foi calculada com base na diferença entre o ano de produção e o ano de início da produção.

Essa variável foi categorizada para representar diferentes estágios do ciclo de vida produtivo:

- Idade < 5 anos → **Novo**
- 5 ≤ Idade < 15 anos → **Maduro**
- Idade ≥ 15 anos → **Antigo**

Essa classificação permite avaliar se a concentração da produção está associada a campos mais recentes ou a ativos mais consolidados.

```python
def classifica_idade(i):
    if i < 5:
        return 'Novo'
    elif i < 15:
        return 'Maduro'
    else:
        return 'Antigo'

df_2025['idade_cat'] = df_2025['idade'].apply(classifica_idade)
```

**Classificação da Profundidade (Lâmina d’Água)**

A lâmina d’água representa a profundidade operacional dos campos offshore. Essa variável foi discretizada para permitir a análise da produção por faixa de profundidade:

- Lâmina d’água < 200 m → **Raso**
- 200 ≤ Lâmina d’água < 1000 m → **Profundo**
- Lâmina d’água ≥ 1000 m → **Ultraprofundo**

Essa categorização é particularmente relevante para o contexto brasileiro, onde a produção em águas profundas e ultraprofundas tem papel central.

```python
def classifica_lamina(m):
    if m < 200:
        return 'Raso'
    elif m < 1000:
        return 'Profundo'
    else:
        return 'Ultraprofundo'

df_2025['lamina_cat'] = df_2025['Lâmina D´Água'].apply(classifica_lamina)
```

## Conclusão

A análise da produção de petróleo no Brasil revela uma dinâmica de concentração mais complexa do que a hipótese inicial sugeria. Embora a expansão do pré-sal esteja associada à entrada em operação de campos de alta produtividade, os resultados indicam que esse movimento não se traduziu em um aumento estrutural da concentração ao longo do tempo. O índice HHI apresenta comportamento cíclico, com tendência de desconcentração até aproximadamente 2015, seguida por um aumento moderado e posterior estabilização, indicando que a expansão do setor foi acompanhada por diversificação relativa entre os campos produtores.

Por outro lado, a análise do período mais recente evidencia que a produção brasileira apresenta forte concentração quando observada sob dimensões estruturais específicas. A predominância de campos em águas ultraprofundas, associada à elevada participação de óleos de grau API médio e à concentração em campos classificados como maduros, indica um alto grau de especialização da indústria.

Esse padrão sugere que a desconcentração observada ao longo do tempo não implica necessariamente maior diversificação estrutural, mas sim uma reorganização da produção dentro de um conjunto relativamente homogêneo de características técnicas e operacionais. Em outras palavras, a produção tornou-se menos concentrada entre campos, mas mais concentrada em termos de perfil produtivo.

Dessa forma, conclui-se que a indústria brasileira de petróleo, no período analisado, combina uma estrutura relativamente menos concentrada do ponto de vista distributivo entre campos com uma forte concentração em termos operacionais, refletindo o papel central do pré-sal na configuração atual do setor.
