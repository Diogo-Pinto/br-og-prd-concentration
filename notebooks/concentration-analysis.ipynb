## Definição do Problema

A produção de petróleo no Brasil passou por mudanças estruturais relevantes nas últimas décadas, especialmente após a entrada em operação dos campos da camada pré-sal a partir de 2010. Esse movimento levanta a necessidade de compreender como a produção está distribuída entre os diferentes campos produtores e se houve aumento na dependência de um número reduzido de ativos de alta relevância.

Diante desse contexto, o problema central deste estudo consiste em avaliar o grau de concentração da produção de petróleo no Brasil ao longo do tempo, buscando identificar se essa produção está distribuída de forma equilibrada entre diversos campos ou concentrada em poucos ativos dominantes.

Além da mensuração direta da concentração por meio do índice de Herfindahl-Hirschman (HHI), o problema é abordado sob diferentes perspectivas estruturais, considerando características dos campos como porte produtivo, localização (onshore/offshore), profundidade operacional, qualidade do petróleo (grau API) e estágio do ciclo de vida produtivo.

Essa abordagem permite não apenas quantificar o nível de concentração, mas também investigar os possíveis fatores associados a esse fenômeno, em linha com as hipóteses levantadas, que relacionam a concentração da produção à entrada de novos campos (pré-sal), ao domínio de ativos de grande porte e a características técnicas e operacionais dos campos produtores.

Trata-se de um problema de análise exploratória de dados, cujo objetivo é identificar padrões estruturais na distribuição da produção e compreender sua evolução ao longo do período analisado (2000–2025).

## Hipóteses do Problema
A entrada em operação dos campos da camada pré-sal, a partir de 2010, aumentou o nível de concentração da produção de petróleo no Brasil, elevando o índice HHI e ampliando a participação relativa dos principais campos produtores.

As hipóteses traçadas são as seguintes:
1. Ao longo do tempo, especialmente após a entrada em operação dos campos do pré-sal, houve um aumento no nível de concentração da produção de petróleo no Brasil, refletido por elevação do índice HHI.

2. A maior parte da produção brasileira está concentrada em um pequeno número de campos de grande porte, definidos com base em sua participação relativa na produção total.

3. A produção nacional é predominantemente proveniente de campos em águas profundas e ultraprofundas.

4. A produção de petróleo no Brasil está majoritariamente associada a campos que produzem óleo de grau API médio a leve.

5. A concentração da produção está associada ao ciclo de vida dos campos, com maior relevância de campos classificados como maduros ou novos.

Para fins de interpretação, serão utilizados os seguintes critérios:

- **Herfindahl-Hirschman Index (HHI)**  
O Índice de Herfindahl-Hirschman (HHI) é uma métrica amplamente utilizada para medir o nível de concentração em um mercado ou, neste caso, na produção de petróleo entre diferentes campos. O HHI é calculado como a soma dos quadrados das participações relativas de cada campo na produção total em um determinado período. Essa abordagem atribui maior peso aos campos com maior participação, tornando a métrica sensível à dominância de poucos ativos. No contexto deste trabalho, o HHI será utilizado para avaliar o grau de concentração da produção ao longo do tempo, permitindo identificar mudanças estruturais na distribuição da produção entre os campos. 
    - HHI < 0.15 → baixa concentração /d
    - 0.15 ≤ HHI < 0.25 → concentração moderada
    - HHI ≥ 0.25 → alta concentração

- **Porte do Campo (Baseado na Participação)**  
A variável de participação percentual foi utilizada para classificar os campos de acordo com sua relevância na produção total mensal.Essa categorização permite identificar se a produção está concentrada em poucos campos de grande porte ou distribuída entre muitos campos menores.
    - Participação < 1% → **Pequeno**
    - 1% ≤ Participação < 5% → **Médio**
    - Participação ≥ 5% → **Grande**

- **Qualidade do Petróleo (Grau API)**
O grau API foi discretizado para classificar o tipo de petróleo produzido, com base em sua densidade e valor econômico.Essa segmentação possibilita analisar se a concentração da produção está associada a determinados tipos de petróleo.
    - API < 22 → **Pesado**
    - 22 ≤ API < 31 → **Médio**
    - 31 ≤ API < 45 → **Leve**
    - API ≥ 45 → **Condensado de Gás**


- **Idade do Campo**
A idade do campo foi calculada a partir da diferença entre o ano de produção e o ano de início da produção, sendo posteriormente categorizada em estágios do ciclo de vida produtivo. Essa classificação permite avaliar se a concentração da produção está associada a campos mais recentes (como os do pré-sal) ou a ativos mais antigos.
    - Idade < 5 anos → **Novo**
    - 5 ≤ Idade < 15 anos → **Maduro**
    - Idade ≥ 15 anos → **Antigo**

- **Profundidade (Lâmina d’Água)**
A lâmina d’água foi discretizada para classificar os campos offshore de acordo com sua profundidade operacional. Essa categorização permite analisar a relação entre profundidade e concentração da produção, especialmente considerando o papel dos campos em águas profundas e ultraprofundas no contexto do pré-sal.
    - Lâmina d’água < 200 m → **Raso**
    - 200 ≤ Lâmina d’água < 1000 m → **Profundo**
    - Lâmina d’água ≥ 1000 m → **Ultraprofundo**
